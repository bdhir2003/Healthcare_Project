```python
from scipy.stats import chi2_contingency
import numpy as np
import pandas as pd

def cramers_v(x, y):
    confusion_matrix = pd.crosstab(x, y)
    chi2 = chi2_contingency(confusion_matrix)[0]
    n = confusion_matrix.sum().sum()
    r, k = confusion_matrix.shape
    return np.sqrt(chi2 / (n * (min(k, r) - 1)))

cat_corr_matrix = pd.DataFrame(index=categorical_features, columns=categorical_features, dtype=float)

for f1 in categorical_features:
    for f2 in categorical_features:
        if f1 == f2:
            cat_corr_matrix.loc[f1, f2] = 1.0
        else:
            cat_corr_matrix.loc[f1, f2] = cramers_v(df[f1], df[f2])

print(cat_corr_matrix)
```

```python
import numpy as np
import pandas as pd

# ========= 1) ICD-9 -> Disease Group mapper =========
def icd9_to_group(code: str) -> str:
    if pd.isna(code) or str(code).strip() in {'', '?'}:
        return 'Unknown'
    s = str(code).strip().upper()
    if s.startswith('V'):
        return 'Supplementary'      # V codes
    if s.startswith('E'):
        return 'ExternalCauses'     # E codes
    # numeric buckets
    try:
        x = float(s)
    except ValueError:
        return 'Unknown'
    if int(x) == 250:                # diabetes explicit
        return 'Diabetes'
    if   1   <= x <= 139:  return 'InfectiousParasitic'
    elif 140 <= x <= 239:  return 'Neoplasms'
    elif 240 <= x <= 279:  return 'EndocrineMetabolic'
    elif 280 <= x <= 289:  return 'Blood'
    elif 290 <= x <= 319:  return 'Mental'
    elif 320 <= x <= 389:  return 'NervousSense'
    elif 390 <= x <= 459:  return 'Circulatory'
    elif x == 785:         return 'Circulatory'
    elif 460 <= x <= 519:  return 'Respiratory'
    elif 520 <= x <= 579:  return 'Digestive'
    elif 580 <= x <= 629:  return 'Genitourinary'
    elif 630 <= x <= 679:  return 'Pregnancy'
    elif 680 <= x <= 709:  return 'Skin'
    elif 710 <= x <= 739:  return 'Musculoskeletal'
    elif 740 <= x <= 759:  return 'Congenital'
    elif 760 <= x <= 779:  return 'Perinatal'
    elif 780 <= x <= 789:  return 'SymptomsSigns'
    elif 790 <= x <= 799:  return 'NonspecificAbnormal'
    elif 800 <= x <= 999:  return 'InjuryPoisoning'
    else:                  return 'Other'

# ========= 2) Main transformer =========
def transform_categoricals(df: pd.DataFrame, topk_med_spec: int = 20) -> pd.DataFrame:
    out = df.copy()

    # --- (A) One-hot low-card categorical features ---
    low_card_cols_desired = [
        'race','gender','age','max_glu_serum','A1Cresult','insulin','change','diabetesMed',
        'admission_type_id','discharge_disposition_id','admission_source_id'
    ]
    # Use only the ones that actually exist
    low_card_cols = [c for c in low_card_cols_desired if c in out.columns]

    # Clean '?' to NaN, then fill for these columns so get_dummies sees categories
    for c in low_card_cols:
        out[c] = out[c].replace('?', np.nan)

    # Cast admin codes to string so theyre treated as categorical (if present)
    for c in ['admission_type_id','discharge_disposition_id','admission_source_id']:
        if c in out.columns:
            out[c] = out[c].astype('Int64').astype(str)

    # get_dummies for low-card set
    out = pd.get_dummies(out, columns=low_card_cols, prefix=low_card_cols, dummy_na=False)

    # --- (B) medical_specialty: group rares -> Other, then one-hot ---
    if 'medical_specialty' in df.columns:
        ms = df['medical_specialty'].replace('?', np.nan).fillna('Unknown')
        topk = ms.value_counts().nlargest(topk_med_spec).index
        ms_grouped = np.where(ms.isin(topk), ms, 'Other')
        ms_dummies = pd.get_dummies(pd.Series(ms_grouped, name='medical_specialty_grouped'),
                                    prefix='medical_specialty')
        # drop any previous medical_specialty columns if present
        out = out.drop(columns=[c for c in out.columns if c.startswith('medical_specialty')], errors='ignore')
        out = pd.concat([out, ms_dummies], axis=1)

    # --- (C) diag_1/diag_2/diag_3: map to disease groups, then one-hot ---
    diag_cols = [c for c in ['diag_1','diag_2','diag_3'] if c in df.columns]
    if diag_cols:
        for dcol in diag_cols:
            grp_col = dcol + '_group'
            out[grp_col] = df[dcol].apply(icd9_to_group)
        # one-hot the group columns
        group_cols = [d + '_group' for d in diag_cols]
        out = pd.get_dummies(out, columns=group_cols, prefix=group_cols, dummy_na=False)
        # (Optional) drop raw diag codes if they also exist in out
        out = out.drop(columns=[c for c in diag_cols if c in out.columns], errors='ignore')

    return out

# ======== 3) Use it ========
# Replace `categorical_df` with the DataFrame that holds your categorical columns
cat_encoded = transform_categoricals(categorical_df, topk_med_spec=20)

print("Original categorical_df shape:", categorical_df.shape)
print("Encoded categorical shape:", cat_encoded.shape)
print("Example columns:", list(cat_encoded.columns)[:15])
```

```python
# --- 1) Target ---
# Convert readmitted into numerical labels: 0 = NO, 1 = >30, 2 = <30
y = cat_encoded['readmitted'].map({'NO': 0, '>30': 1, '<30': 2})

# --- 2) Features ---
# Drop the target column from categorical features
X_categorical = cat_encoded.drop(columns=['readmitted'])

# Make sure both categorical and numerical dfs are aligned by index
X_categorical = X_categorical.loc[normalized_numerical_df.index]

# Concatenate numerical + categorical features
X = pd.concat([normalized_numerical_df, X_categorical], axis=1)

print(" Final shapes")
print("X shape:", X.shape)      # should be (101766, 168 + numeric cols)
print("y shape:", y.shape)      # should be (101766,)
print("Target distribution:", y.value_counts(normalize=True).round(3).to_dict())
```

```python
from imblearn.over_sampling import SMOTE

# Only oversample <30 days
sm = SMOTE(sampling_strategy={2: y.value_counts()[1]}, random_state=42)
X_res, y_res = sm.fit_resample(X, y)

print("Original class counts:", y.value_counts().to_dict())
print("Resampled class counts:", y_res.value_counts().to_dict())
# Pie chart of class distribution before SMOTE
plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1)
y.value_counts().plot.pie(autopct='%1.1f%%', labels=['NO', '>30', '<30'], colors=['#66b3ff','#99ff99','#ff9999'])
plt.title('Class Distribution Before SMOTE')
plt.ylabel('')

# Pie chart of class distribution after SMOTE
plt.subplot(1, 2, 2)
y_res.value_counts().plot.pie(autopct='%1.1f%%', labels=['NO', '>30', '<30'], colors=['#66b3ff','#99ff99','#ff9999'])
plt.title('Class Distribution After SMOTE')
plt.ylabel('')

plt.tight_layout()
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, accuracy_score, ConfusionMatrixDisplay
import matplotlib.pyplot as plt

# Split resampled dataset
X_train, X_test, y_train, y_test = train_test_split(
    X_res, y_res, test_size=0.2, stratify=y_res, random_state=42
)

# Train RF
rf = RandomForestClassifier(
    n_estimators=600,
    max_depth=None,
    n_jobs=-1,
    random_state=42
)
rf.fit(X_train, y_train)

# Predictions
y_pred = rf.predict(X_test)

# Evaluation
print("\n=== Random Forest (SMOTE oversampled) ===")
print("Accuracy:", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred, digits=3))

ConfusionMatrixDisplay.from_predictions(y_test, y_pred)
plt.title("Random Forest  SMOTE Balanced Data")
plt.tight_layout()
plt.show()
```

```python
import re
from xgboost import XGBClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report, ConfusionMatrixDisplay
import matplotlib.pyplot as plt

def sanitize_cols(df):
    cols, seen = [], {}
    for c in df.columns.astype(str):
        s = re.sub(r'[^0-9a-zA-Z_]+', '_', c)   # replace bad chars with _
        s = re.sub(r'_+', '_', s).strip('_')    # collapse repeats
        if s in seen:
            seen[s] += 1
            s = f"{s}_{seen[s]}"                # ensure uniqueness
        else:
            seen[s] = 0
        cols.append(s)
    out = df.copy()
    out.columns = cols
    return out

# Apply to the SMOTE-resampled features BEFORE splitting
X_res_clean = sanitize_cols(X_res)

X_train, X_test, y_train, y_test = train_test_split(
    X_res_clean, y_res, test_size=0.2, stratify=y_res, random_state=42
)

xgb = XGBClassifier(
    objective='multi:softprob', num_class=3,
    n_estimators=800, learning_rate=0.08, max_depth=6,
    subsample=0.9, colsample_bytree=0.8,
    tree_method='hist', eval_metric='mlogloss', random_state=42, n_jobs=-1
)

xgb.fit(X_train, y_train)
y_pred = xgb.predict(X_test)

print("Accuracy:", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred, digits=3))
ConfusionMatrixDisplay.from_predictions(y_test, y_pred)
plt.title("XGBoost  SMOTE balanced (sanitized cols)")
plt.tight_layout(); plt.show()
```

```python
# Analyze XGBoost model performance across race, gender, and age (proxy for socioeconomic group)

# 1. Get test set indices and predictions from XGBoost cell
test_idx = y_test.index
y_true_xgb = y_test
y_pred_xgb = y_pred  # from XGBoost cell

# 2. Get demographic info for test set
demog_cols = ['race', 'gender', 'age']
demog_test = df.loc[test_idx, demog_cols].reset_index(drop=True)

# 3. Combine for analysis
results_xgb = pd.DataFrame({
    'y_true': y_true_xgb.values,
    'y_pred': y_pred_xgb,
    'race': demog_test['race'],
    'gender': demog_test['gender'],
    'age': demog_test['age']
})

# 4. Function to compute metrics per group
def group_metrics(df, group_col):
    metrics = []
    for group, subdf in df.groupby(group_col):
        acc = np.mean(subdf['y_true'] == subdf['y_pred'])
        fp = np.sum((subdf['y_true'] != 2) & (subdf['y_pred'] == 2)) / len(subdf)
        fn = np.sum((subdf['y_true'] == 2) & (subdf['y_pred'] != 2)) / len(subdf)
        metrics.append({'group': group, 'n': len(subdf), 'accuracy': acc, 'false_pos_rate': fp, 'false_neg_rate': fn})
    return pd.DataFrame(metrics).sort_values('group')

# 5. Race
print("\n=== XGBoost Performance by Race ===")
print(group_metrics(results_xgb, 'race'))

# 6. Gender
print("\n=== XGBoost Performance by Gender ===")
print(group_metrics(results_xgb, 'gender'))

# 7. Age (bucketed, as a proxy for socioeconomic group)
results_xgb['age_bucket'] = results_xgb['age'].str.extract(r'\[(\d+-\d+)\)')
print("\n=== XGBoost Performance by Age Bucket ===")
print(group_metrics(results_xgb, 'age_bucket'))
```

```python
from tensorflow import keras
from tensorflow.keras import layers
from sklearn.model_selection import train_test_split
import numpy as np

X_train, X_test, y_train, y_test = train_test_split(
    X_res_num.values, y_res.values, test_size=0.2, stratify=y_res, random_state=42
)

num_classes = len(np.unique(y_res))
y_train_cat = keras.utils.to_categorical(y_train, num_classes)
y_test_cat  = keras.utils.to_categorical(y_test,  num_classes)

model = keras.Sequential([
    layers.Input(shape=(X_train.shape[1],)),
    layers.Dense(256, activation="relu"),
    layers.Dropout(0.3),
    layers.Dense(128, activation="relu"),
    layers.Dropout(0.3),
    layers.Dense(64, activation="relu"),
    layers.Dense(num_classes, activation="softmax")
])

model.compile(optimizer=keras.optimizers.Adam(1e-3),
              loss="categorical_crossentropy",
              metrics=["accuracy"])

history = model.fit(
    X_train, y_train_cat,
    validation_split=0.2,
    epochs=20,
    batch_size=256,
    verbose=1
)

test_loss, test_acc = model.evaluate(X_test, y_test_cat, verbose=0)
print(f"\nDNN Test Accuracy: {test_acc:.3f}")

y_pred = model.predict(X_test)
y_pred = np.argmax(y_pred, axis=1)
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred, digits=3))
```
